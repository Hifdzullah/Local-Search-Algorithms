```python
import random
import string
import time
import copy

N = input("Enter N: ")
N = int(N)

start_time = time.time()
def random_board(N):
       return [ random.randint(1, N) for _ in range(N) ]
board = random_board(N)
print ("Initial board: ",board)

#newline

def heuristic_value(individual):
    horizontal_collisions = sum([individual.count(queen)-1 for queen in individual])/2
    diagonal_collisions = 0

    n = len(individual)
    left_diagonal = [0] * 2*n
    right_diagonal = [0] * 2*n
    for i in range(n):
        left_diagonal[i + individual[i] - 1] += 1
        right_diagonal[len(individual) - i + individual[i] - 2] += 1

    diagonal_collisions = 0
    for i in range(2*n-1):
        counter = 0
        if left_diagonal[i] > 1:
            counter += left_diagonal[i]-1
        if right_diagonal[i] > 1:
            counter += right_diagonal[i]-1
        diagonal_collisions += counter #/ (n-abs(i-n+1))
    
    return int(horizontal_collisions + diagonal_collisions)

def hill_climbing(board):
    # Find the least cost successor for the given board state

    min_board = board
    min_h = 999999
    global n_side_moves, n_steps

    n_steps += 1

    # Check if number of side moves has reached a limit
    if n_side_moves == 100:
        return min_board

    sideway_move = False

    for i in range(N):
        # Find index of queen in current row
        queen = board[i]
        k=0
        # for k in range (N) and k != (board[i]-1):
        for k in range (N):
            if k != board[i]-1:
                board[i] = k+1

                h = heuristic_value(board)

                if h < min_h:
                    min_h = h
                    min_board = copy.deepcopy(board)
                if h == min_h:
                    min_h = h
                    min_board = copy.deepcopy(board)
                    sideway_move = True
        board[i] = queen

    if sideway_move:
        n_side_moves += 1

    if min_h == 0:
        
        return min_board

    return hill_climbing(min_board)

def random_restart(min_board):
    
    maxrestart = 100
    
    for _ in range(maxrestart):
            
        global n_of_restarts, n_side_moves, n_steps
        n_of_restarts += 1
        new_board = random_board(N)
        hill_climbing(new_board)

        if heuristic_value(new_board) < heuristic_value(min_board):

            min_board = hill_climbing(new_board)
            
        elif heuristic_value(new_board) == 0:
            
            min_board = hill_climbing(new_board)
            break
        
        n_side_moves = 0
        n_steps = 0
        
    return min_board
    
n_of_restarts = 0
n_side_moves = 0
n_steps = 0

#print("Current position's heuristic value: ", heuristic_value(board))

min_board = hill_climbing(board)

if n_side_moves == 100:
        
        min_board = random_restart(min_board)

if heuristic_value(min_board) == 0:
    
        print("Final board: ",min_board)
        print("Number of restarts: ", n_of_restarts)
        print("Final Heuristic value: ",heuristic_value(min_board))
    
else:
    print("Could not solve")
    print("Final board: ", min_board)
    print("Number of restarts: ", n_of_restarts)
    print("Final Heuristic value: ",heuristic_value(min_board))

print("Time taken: ""%s seconds" % (time.time() - start_time))
```
